# Polynomial Regression

Here we are going to build a regression model for a non linear dataset.

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('/content/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```
print(X)
```

What is this dataset about?
Lets imagine you are actually working in an HR department and you want to hire someone.
And you actually found someone who seem to be a great fit for the job.
At the end of the Interview process comes the question what is your salary  expectation?
So lets assume that this person is asking for 160k dollars per year.
Then as HR you ask this person why are you expecting such high salary.
And this person says well thats because thats what i used to earn in my previous company.
Is that a truth or bluff?
To find out lets build a polynomial linear regression model.
To make such position we need data.










## Training the linear regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

## Training the Polynomial Regression model on the whole dataset

regression.png

As you can see these are the equations for the models.
For linear and multiple linear model you can see that the output depends on x0 x1 etc.
While in polynomial regression equation u could see that the ouput depends on increasing powers of x1. For that we need to create a matrix containing the powers of x1. We will then create a linear regressor object and then integrate them together.

First step is to import the tools required for making the matrix

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2) # first we will try degree 2 ie upto x1 square then we will try 3 and 4
X_poly=poly_reg.fit_transform(X) #by doing this we will have new matrix of features
#now we have new set of features
#now we have to build a linear regression model to integrate these features
```

```
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

## Visualising the linear regression results

```
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg.predict(X),color='blue')
plt.title('Truth Or Bluff')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

So this is the linear regression result.
You can see that its not a good model.The predictions are far from the real values

## Visualising the polynomial regression results

```
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_2.predict(X_poly),color='blue')
plt.title('Truth Or Bluff')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

So this is much better.
This is the result for degree 2.
Higher degrees would result in much better results.
Lets do that. First lets do that for degree 2.

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=3) # we will try 3 and 4
X_poly=poly_reg.fit_transform(X) #by doing this we will have new matrix of features
#now we have new set of features
#now we have to build a linear regression model to integrate these features
```

```
lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly,y)
```

```
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_3.predict(X_poly),color='blue')
plt.title('Truth Or Bluff')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

See we get a much better result than degree 2.
Now lets do for degree 4

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4) # we will try 3 and 4
X_poly=poly_reg.fit_transform(X) #by doing this we will have new matrix of features
#now we have new set of features
#now we have to build a linear regression model to integrate these features
```

```
lin_reg_4 = LinearRegression()
lin_reg_4.fit(X_poly,y)
```

```
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_4.predict(X_poly),color='blue')
plt.title('Truth Or Bluff')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

See we get an even better result here

## Predicting a new result with Linear Regression

```
lin_reg.predict([[6.5]])
```

## Predicting a New result with polynomial regression

```
lin_reg_4.predict(poly_reg.fit_transform([[6.5]]))
```

## Model parameters

1.   coeffcients
2.   Intercept



```
print("Coeficients",lin_reg_4.coef_)
print("Intercept",lin_reg_4.intercept_)
```
